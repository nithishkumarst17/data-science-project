**Amazon Sales**

```python
import numpy as np
import pandas as pd

np.random.seed(42)

num_products = 2000
products = [f"Product_{i+1}" for i in range(num_products)]

data = []

for product in products:
    ad_spend_flag = np.random.choice(['Yes', 'No'], p=[0.7, 0.3])
    monthly_sales = np.random.randint(1000, 5000, 6)  # last 6 months sales

    if ad_spend_flag == 'Yes':
        improvement_factor = np.random.uniform(1.05, 1.3)  # sales increase due to ad
        sales_after_ad = int(monthly_sales[-1] * improvement_factor)
    else:
        improvement_factor = 1
        sales_after_ad = monthly_sales[-1]

    sales_before_ad = monthly_sales[-1]
    sales_percentage_change = round(((sales_after_ad - sales_before_ad) / sales_before_ad) * 100, 2)
    plan_success = 'Yes' if sales_after_ad > sales_before_ad else 'No'

    data.append([product, list(monthly_sales), ad_spend_flag, sales_before_ad, sales_after_ad, sales_percentage_change, plan_success])

df_summary = pd.DataFrame(data, columns=[
    'Product', 'Last_6_Months_Sales', 'Advertisement', 'Sales_Before_Ad', 'Sales_After_Ad',
    'Sales_Improvement_Percent', 'Plan_Success'
])

print(df_summary.head(10))
```
