```python
#  generate code for  univariate analysis

import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# Univariate Analysis for Numerical Features
for column in data.select_dtypes(include=np.number):
  print(f"Univariate Analysis for '{column}'")
  print(data[column].describe())
  plt.figure(figsize=(8, 6))
  sns.histplot(data[column], kde=True)
  plt.title(f"Distribution of {column}")
  plt.show()

  print("\n")
```

```python
# Univariate Analysis for Categorical Features
for column in data.select_dtypes(include='object'):
  print(f"Univariate Analysis for '{column}'")
  print(data[column].value_counts())
  plt.figure(figsize=(8, 6))
  sns.countplot(x=data[column])
  plt.title(f"Countplot of {column}")
  plt.xticks(rotation=45, ha='right')
  plt.show()
  print("\n")
```

***Donut Chart for Embarked Status***

```python
import matplotlib.pyplot as plt
import seaborn as sns

colors = sns.color_palette("muted")
# Count the occurrences of each embarking ports for travel
company_counts = data['Embarked'].value_counts()
# Create a donut chart
plt.figure(figsize=(8, 8))
plt.pie(company_counts, labels=company_counts.index, autopct='%1.1f%%', startangle=90, colors=colors)
my_circle = plt.Circle((0, 0), 0.3, color='white')
p = plt.gcf()
p.gca().add_artist(my_circle)
plt.title('Distribution of travellers based on their embarked status')
plt.axis('equal')  # Equal aspect ratio ensures that the centre is drawn as a circle.
plt.show()
```
